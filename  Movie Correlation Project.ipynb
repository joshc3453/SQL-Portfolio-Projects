```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Reading in the data
df = pd.read_csv(r'C:\Users\joshc\Downloads\movies.csv')
```

```python
# Looking at data
df.head()
```

```python
# Identify any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data types for columns

df.dtypes
```

```python
# Change data type of columns
df['budget'] = df['budget'].astype('Int64')
```

```python
# Change data type of columns
df['gross'] = df['gross'].astype('Int64')
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Creating correct Year column (There seems to be some discrepancies between the "released" column year and the "year" column)

NewDataFrame = df['released'].str.split(" ",n=3,expand = True) 
```

```python
NewDataFrame.rename(columns={0: 'Month', 1: 'day',2:'year',3:'country'}, inplace=True)
```

```python
df['yearcorrect'] = NewDataFrame['year']
```

```python
df['released'].str.split(" ",n=3,expand = True).head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
df['gross'] = df['gross'].astype('float')
df['budget'] = df['budget'].astype('float')
```

```python
# Exploring correlation between Budget and Gross Revenue
```

```python
# Scatter plot: Budget vs. Gross Revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Company Data

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
