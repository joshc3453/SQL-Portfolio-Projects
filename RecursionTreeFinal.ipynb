```python
import ColabTurtle.Turtle as turtle

turtle.initializeTurtle()
turtle.bgcolor('deepskyblue')

def drawTree(num_branches, branch_length, angle):
    if num_branches > 0:
        turtle.forward(branch_length)
        turtle.left(angle)
        drawTree(num_branches-1, branch_length * 0.75, angle)
        
        turtle.right(2*angle)
        drawTree(num_branches - 1, branch_length * 0.75, angle)
        
        turtle.left(angle)
        turtle.forward(-branch_length)

def drawSun(sun_branches, sun_length, sun_angle):
    if sun_branches > 0:
        turtle.forward(sun_length)
        turtle.left(sun_angle)
        drawSun(sun_branches-1, sun_length * 0.99, sun_angle)

def drawBird(num_branches, branch_length, angle):
    if num_branches > 0:
        turtle.forward(branch_length)
        turtle.right(angle)
        drawBird(num_branches-1, branch_length * 0.75, angle)
                
        turtle.left(angle)
        turtle.forward(-branch_length)

def drawBird2(num_branches, branch_length, angle):
    if num_branches > 0:
        turtle.forward(branch_length)
        turtle.left(angle)
        drawBird2(num_branches-1, branch_length * 0.75, angle)
        
        turtle.right(angle)
        turtle.forward(-branch_length)

turtle.speed(10)
turtle.up()

num_branches = 5
branch_length = 50
angle = 35
start_coords = (100, 500)

turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('forestgreen')
drawTree(num_branches, branch_length, angle)

start_coords = (200, 500)
turtle.up()
turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('darkgreen')
drawTree(num_branches, branch_length, angle)

start_coords = (turtle.window_width()/2, turtle.window_height())
turtle.up()
turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('darkolivegreen')
drawTree(6, 80, angle)

start_coords = (turtle.window_width()-200, turtle.window_height())
turtle.up()
turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('green')
drawTree(num_branches, branch_length, angle)

start_coords = (turtle.window_width()-100, turtle.window_height())
turtle.up()
turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('olivedrab')
drawTree(num_branches, branch_length, angle)

start_coords = (turtle.window_width()-100, turtle.window_height()-375)
turtle.up()
turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('yellow')
drawSun(140, 50, 35)

num_branches = 3
branch_length = 20
angle = 60

start_coords = (turtle.window_width()-400, turtle.window_height()-300)
turtle.up()
turtle.setpos(start_coords)
turtle.down()
turtle.pencolor('black')
turtle.setheading(270)
drawBird(num_branches, branch_length, angle)
drawBird2(num_branches, branch_length, angle)

start_coords = (turtle.window_width()-370, turtle.window_height()-350)
turtle.up()
turtle.setpos(start_coords)
turtle.down()
drawBird(num_branches, branch_length, angle)
drawBird2(num_branches, branch_length, angle)

start_coords = (turtle.window_width()-430, turtle.window_height()-350)
turtle.up()
turtle.setpos(start_coords)
turtle.down()
drawBird(num_branches, branch_length, angle)
drawBird2(num_branches, branch_length, angle)
turtle.hideturtle()
```
